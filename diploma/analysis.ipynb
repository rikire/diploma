```python
# Подготовка датасета для регрессии временного ряда (целевой признак - counts)
# 1. Сортировка по дате
daily_sunspots_df = daily_sunspots_df.sort_values(by=['year', 'month', 'day']).reset_index(drop=True)

# 2. Заполнение пропусков и преобразование признаков
daily_sunspots_df = daily_sunspots_df.fillna(0)
daily_sunspots_df['indicator'] = daily_sunspots_df['indicator'].replace('*', 1).replace(0, 0)

# 3. Формирование признаков для регрессии временного ряда (лаги)
def create_lag_features(df, target_col, lags=7):
    for lag in range(1, lags + 1):
        df[f'{target_col}_lag_{lag}'] = df[target_col].shift(lag)
    return df

daily_sunspots_df = create_lag_features(daily_sunspots_df, 'counts', lags=7)

# 4. Удаление строк с NaN после создания лагов
daily_sunspots_df = daily_sunspots_df.dropna().reset_index(drop=True)

# 5. Выбор признаков для обучения
features = [f'counts_lag_{i}' for i in range(1, 8)]
X = daily_sunspots_df[features]
y = daily_sunspots_df['counts']
```

```python
daily_sunspots_lags = X.copy()
daily_sunspots_lags['counts'] = y.values
daily_sunspots_lags.to_csv('daily_sunspots_lags.csv', index=False, encoding='utf-8')
```

```python
import h2o
from h2o.automl import H2OAutoML

# Инициализация H2O
h2o.init()

# Загрузка данных
data = h2o.import_file("/mnt/c/study/paper/diploma/datasets/daily_sunspots_lags.csv")

# Преобразование целевой переменной в категориальный тип (если задача классификации)
data['counts'] = data['counts'].asfactor()

# Разделение на обучающую и тестовую выборки
train, test = data.split_frame(ratios=[0.8], seed=123)

# Определение признаков и целевой переменной
x = [col for col in data.columns if col != 'counts']
y = 'counts'

# Создание и обучение модели
aml = H2OAutoML(max_models=20, max_runtime_secs=600, seed=1)
aml.train(x=x, y=y, training_frame=train)

# Получение лучшей модели
leader = aml.leader

# Оценка на тестовой выборке
perf = leader.model_performance(test_data=test)
print(perf)

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Загрузка данных
df = pd.read_csv("путь_к_вашему_датасету.csv")

# Разделение на признаки и целевую переменную
X = df.drop('Целевая_переменная', axis=1)
y = df['Целевая_переменная']

# Преобразование целевой переменной в числовой формат (если необходимо)
le = LabelEncoder()
y = le.fit_transform(y)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

import autokeras as ak

# Создание и обучение модели
clf = ak.StructuredDataClassifier(max_trials=20, overwrite=True)
clf.fit(X_train, y_train, epochs=10)

# Оценка на тестовой выборке
loss, acc = clf.evaluate(X_test, y_test)
print(f"Точность модели AutoKeras: {acc:.3f}")

```
